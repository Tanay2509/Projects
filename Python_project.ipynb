```python
data = open("C:/Users/ASUS/Desktop/Python Project/conv.txt", encoding = 'utf-8')
dia = ''
for i in data:
 dia += i  #dia is the complete text file in string format

x = dia.split('\n\n') #splitting text file into a list

#splitting the list x into two lists. First list containing name of all the chars and second containing all the dialogues. 
y = []
for i in range(len(x)):
  element = x[i].split(': ')
  y.append(element)
chars = []
dialogues = []
for i in range(len(y)-1):
  chars.append(y[i][0])
  dialogues.append(y[i][1])
```

```python
#uncomment to see the result
#dia
#x
#y
#chars
#dialogues
#making a set of list 'chars' to remove duplicate chars
unique_chars = set(chars)
unique_chars_list = list(unique_chars)
#unique_chars_list
len(unique_chars)
```

```python
#print(y)
def dialogues_by(name):
    words = ''
    for i in range(len(y)):
        if name in y[i]:
            words+= (y[i][1].replace('.','').replace(',','').replace('?','')).replace('…','').replace('!','') #replacing punctuations from dialogues
            words+= ' ' #concatenating dialogues 
            words = words.lower()
            word_list = words.split(' ')
            word_list.remove('') #removing empty spaces
            unique_words_set = set(word_list) #creating a set to remove duplicate words
            unique_words_list = list(unique_words_set)
            unq=[i for i in unique_words_list if i != '']
    return unq
    
        
    
        
```

```python
for i in range(len(unique_chars)):
    f_name='C:/Users/ASUS/Desktop/Python Project/files/'+unique_chars_list[i]+'.txt'
    file = open(f_name,'a',encoding = 'utf-8')
    a= dialogues_by(unique_chars_list[i])
    for words in a:
        b=words+'\n'
        file.write(b)
    file.close()
data.close()
```
